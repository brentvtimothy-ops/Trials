```
# Make the largest possible number from an array of non-negative integers, using a custom string comparator.

def largest_number(nums):
    # Convert integers to strings
    strs = []
    for num in nums:
        strs.append(str(num))
    
    # Custom comparator: returns True if a should come before b
    def comes_before(a, b):
        return a + b > b + a
    
    # Merge sort implementation
    def merge_sort(arr):
        if len(arr) <= 1:
            return arr
        
        mid = len(arr) // 2
        left = merge_sort(arr[:mid])
        right = merge_sort(arr[mid:])
        
        return merge(left, right)
    
    def merge(left, right):
        result = []
        i = j = 0
        
        while i < len(left) and j < len(right):
            if comes_before(left[i], right[j]):
                result.append(left[i])
                i += 1
            else:
                result.append(right[j])
                j += 1
        
        # Append remaining elements
        while i < len(left):
            result.append(left[i])
            i += 1
        
        while j < len(right):
            result.append(right[j])
            j += 1
        
        return result
    
    # Sort using custom merge sort
    sorted_strs = merge_sort(strs)
    
    # Handle edge case: all zeros
    if sorted_strs[0] == "0":
        return "0"
    
    # Concatenate result
    result = ""
    for s in sorted_strs:
        result += s
    
    return result

# Example usage
#nums = [3, 30, 34, 5, 9]
#nums = [1, 61, 9, 0]
#nums = [0,0,0,1,56,71,987,12,114,23,73,24]
nums = [9,98,9,99]
print(largest_number(nums))  # Output: "9534330"
```

```
# Make the largest possible number from an array of non-negative integers, using a custom string comparator.
# more memory-efficient version, but can be less stable
def largest_number(nums):
    # Convert integers to strings
    strs = [str(num) for num in nums]
    
    # Comparator: True if a should come before b
    def comes_before(a, b):
        return a + b > b + a
    
    # In-place quicksort  --instead of merge sort
        # The quick sort is internal sorting method where the data is sorted in main memory. 
        #The merge sort is external sorting method in which the data that is to be sorted cannot
            #be accommodated in the memory and needed auxiliary memory for sorting.
        #Merge sort is stable as two elements with equal value appear in the same order in 
            #sorted output as they were in the input unsorted array. 
        #Quick sort is not stable in this scenario. 
            #But it can be made stable using some changes in code.
        #Quick sort is preferred for arrays.
        #Merge sort is preferred for linked lists. 
        #Quick Sort performs better in general, but Merge Sort works better for external sorting.
    
    def quicksort(arr, low, high):
        if low < high:
            pivot_index = partition(arr, low, high)
            quicksort(arr, low, pivot_index - 1)
            quicksort(arr, pivot_index + 1, high)
    
    def partition(arr, low, high):
        pivot = arr[high]
        i = low
        
        for j in range(low, high):
            if comes_before(arr[j], pivot):
                arr[i], arr[j] = arr[j], arr[i]
                i += 1
        
        # Place pivot in correct position
        arr[i], arr[high] = arr[high], arr[i]
        return i
    
    # Sort the array
    quicksort(strs, 0, len(strs) - 1)
    
    # Edge case: if the largest is "0", return "0"
    if strs and strs[0] == "0":
        return "0"
    
    # Concatenate result
    result = ""
    for s in strs:
        result += s
    
    return result

# Example usage
nums = [3, 30, 34, 5, 9]
print(largest_number(nums))  # Output: "9534330"
```

```
# Find all pairs that sum to a target in an array with positive and negative numbers, 
    #using hashing and two pointers.
def find_pairs_hash(nums, target):
    seen = {}
    pairs = {}

    for num in nums:
        complement = target - num

        if complement in seen:
            # Normalize pair order to avoid duplicates like (2,3) and (3,2)
            a = min(num, complement)
            b = max(num, complement)
            pairs[(a, b)] = True

        seen[num] = True

    # Convert keys to list of tuples
    result = []
    for pair in pairs:
        result.append(pair)

    return result


# Example
# nums = [1, 2, -1, 0, -2, 1, -1, 3]
nums = [1,1,2,2,3,4,5]
target = 4
print(find_pairs_hash(nums, target))
```

```
#check string segmentation using a given dictionary, recursion with memoization, or DP
def word_break_memo(s, word_dict):
    # Edge cases
    if s is None:
        return False
    if s == "":
        return True

    word_set = set(word_dict)
    memo = {}

    # Precompute max word length for pruning
    max_len = 0
    for w in word_set:
        if len(w) > max_len:
            max_len = len(w)

    def dfs(start):
        if start == len(s):
            return True

        if start in memo:
            return memo[start]

        # Try all possible word lengths
        for end in range(start + 1, min(len(s) + 1, start + max_len + 1)):
            if s[start:end] in word_set and dfs(end):
                memo[start] = True
                return True

        memo[start] = False
        return False

    return dfs(0)


# Example
#s = "leetcode"
#word_dict = ["leet", "code"]
s = "batcricketbat"
word_dict = ["bat", "cricket"]
print(word_break_memo(s, word_dict))  # True
```

```
# Find the missing number in an array from range 0 to n , using the sum or XOR trick.
    #Sum method can cause overflow
def find_missing_xor(nums):
    n = len(nums)

    xor_all = 0
    xor_arr = 0

    seen = {}

    # XOR all indices and values
    for i in range(n):
        xor_all ^= i
        xor_arr ^= nums[i]

        # Validate duplicates and range
        if nums[i] < 0 or nums[i] > n:
            raise ValueError("Value out of valid range [0, n]")
        if nums[i] in seen:
            raise ValueError("Duplicate value detected")
        seen[nums[i]] = True

    xor_all ^= n  # include n

    return xor_all ^ xor_arr


def find_missing_sum(nums):
    n = len(nums)

    expected_sum = n * (n + 1) // 2
    actual_sum = 0

    seen = {}

    for num in nums:
        # Validate range
        if num < 0 or num > n:
            raise ValueError("Value out of valid range [0, n]")
        
        # Detect duplicates
        if num in seen:
            raise ValueError("Duplicate value detected")
        seen[num] = True

        actual_sum += num

    return expected_sum - actual_sum

# Example
#nums = [3, 0, 1]
nums = [5,2,3,1,0]
print(find_missing_xor(nums))  # Output: 2
print(find_missing_sum(nums)) 
```

```
#Print Fibonacci sequence using recursion, recursion basics, and tail recursion
max_n = 1000
#Basic
def fib_basic(n):
    validate_n(n)
    if n <= 1:
        return n
    return fib_basic(n - 1) + fib_basic(n - 2)

def print_fib_basic(n):
    validate_n(n)
    #result = []
    #for i in range(n):
    #    result.append(fib_basic(i)) #full series to n-1
    result=fib_basic(n) #just last in series
    return result

def fib_memo(n, memo):
    if n in memo:
        return memo[n]
    
    memo[n] = fib_memo(n - 1, memo) + fib_memo(n - 2, memo)
    return memo[n]

#Recursion + Memoization
def print_fib_memo(n):
    validate_n(n)
    
    if n == 0:
        return []
    
    memo = {0: 0, 1: 1}
    result = []
    
    for i in range(n):
        result.append(fib_memo(i, memo))
    
    return result

# Tail Recursion Style    
def fib_tail(n, a=0, b=1, result=None):
    validate_n(n)
    
    if result is None:
        result = []
    
    if n == 0:
        return result
    
    result.append(a)
    return fib_tail(n - 1, b, a + b, result)

def validate_n(n, max_n=1000):
    if n < 0:
        raise ValueError("n must be non-negative")
    if n > max_n:
        raise ValueError("n too large for recursion; use iterative approach")

# Example
print(fib_tail(11))
print(print_fib_memo(11))
print(print_fib_basic(10))
```
